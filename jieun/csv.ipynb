### **ccmedia tar unzip**

```python
import os
import zipfile

def extract_all_zip_files(root_dir):
    # 모든 하위 디렉토리와 파일을 순회
    for subdir, _, files in os.walk(root_dir):
        for file in files:
            if file.endswith(".tar"):
                file_path = os.path.join(subdir, file)
                # 파일 이름과 동일한 디렉토리를 생성
                extract_dir = os.path.join(subdir, os.path.splitext(file)[0])
                
                # 디렉토리가 이미 존재하지 않으면 생성
                if not os.path.exists(extract_dir):
                    os.makedirs(extract_dir)
                
                try:
                    # zip 파일 열기
                    with zipfile.ZipFile(file_path, 'r') as zip_ref:
                        print(f"Extracting {file_path} to {extract_dir}")
                        # 압축 해제
                        zip_ref.extractall(extract_dir)
                except zipfile.BadZipFile as e:
                    print(f"Failed to extract {file_path}: {e}")
                except Exception as e:
                    print(f"An error occurred with {file_path}: {e}")
```

### **AIHub tar unzip**

```python
import os
import tarfile

# 압축을 풀 루트 디렉토리 경로
root_dir = '/mnt/aix7206/data/old'  # 여기에 루트 디렉토리 경로 입력

# 모든 하위 디렉토리를 순회하며 .tar 파일 압축 해제
for dirpath, dirnames, filenames in os.walk(root_dir):
    for filename in filenames:
        if filename.endswith('.tar'):  # .tar 파일인지 확인
            file_path = os.path.join(dirpath, filename)
            print(f'압축 해제 중: {file_path}')

            # 압축 해제
            try:
                with tarfile.open(file_path) as tar:
                    tar.extractall(path=dirpath)  # 해당 디렉토리에 압축 해제
                print(f'{file_path} 압축 해제 완료!')
            except Exception as e:
                print(f'압축 해제 실패: {file_path} - {str(e)}')
```

### **json to csv**

```python
def combine_json_files_recursive(directory):
    combined_df = pd.DataFrame()

    for root, _, files in os.walk(directory):
        for file in files:
            if file.endswith('.json'):
                file_path = os.path.join(root, file)
                with open(file_path, 'r') as f:
                    data = json.load(f)
                    
                    # 필요한 데이터 추출
                    image_info = data.get('images', {})
                    environments = data.get('environments', [])
                    growth_index = data.get('growth_index', {})
                    etc_infor = data.get('etc_infor', {})
                    annotations = data.get('annotations', [])
                    licenses = data.get('licenses', [])
                    categories = data.get('categories', [])
                    
                    # 데이터 프레임 생성
                    image_df = pd.json_normalize(image_info).add_prefix('image_')
                    env_df = pd.json_normalize(environments).add_prefix('env_')
                    growth_df = pd.json_normalize(growth_index).add_prefix('growth_')
                    etc_df = pd.json_normalize(etc_infor).add_prefix('etc_')
                    ann_df = pd.json_normalize(annotations).add_prefix('ann_')
                    lic_df = pd.json_normalize(licenses).add_prefix('lic_')
                    cat_df = pd.json_normalize(categories).add_prefix('cat_')

                    # 데이터 프레임 병합
                    merged_df = pd.concat([image_df, env_df, growth_df, etc_df, ann_df, lic_df, cat_df], axis=1)
                    
                    # 전체 데이터 프레임에 추가
                    combined_df = pd.concat([combined_df, merged_df], ignore_index=True)

    return combined_df
```

# ***2024.09.25 after data updates***

```python
def combine_json_files_recursive(directory):
    combined_df = pd.DataFrame()

    for root, _, files in os.walk(directory):
        for file in files:
            if file.endswith('.json'):
                file_path = os.path.join(root, file)
                with open(file_path, 'r', encoding='utf-8') as f:
                    data = json.load(f)
                    
                    # 필요한 데이터 추출
                    image_info = data.get('images', {})
                    environments = data.get('environments', [])
                    growth_index = data.get('growth_index', {})
                    etc_infor = data.get('etc_infor', {})
                    annotations = data.get('annotations', [])
                    licenses = data.get('licenses', [])
                    categories = data.get('categories', [])
                    
                    # 이미지 정보
                    image_df = pd.json_normalize(image_info).add_prefix('image_') if image_info else pd.DataFrame()
                    # 환경 정보
                    env_df = pd.json_normalize(environments).add_prefix('env_') if environments else pd.DataFrame()
                    # 성장 지수
                    growth_df = pd.json_normalize(growth_index).add_prefix('growth_') if growth_index else pd.DataFrame()
                    # 기타 정보
                    etc_df = pd.json_normalize(etc_infor).add_prefix('etc_') if etc_infor else pd.DataFrame()
                    # 어노테이션 정보
                    ann_df = pd.json_normalize(annotations).add_prefix('ann_') if annotations else pd.DataFrame()
                    # 라이선스 정보
                    lic_df = pd.json_normalize(licenses).add_prefix('lic_') if licenses else pd.DataFrame()
                    # 카테고리 정보
                    cat_df = pd.json_normalize(categories).add_prefix('cat_') if categories else pd.DataFrame()

                    # 데이터 프레임 병합 (여러 어노테이션이 있을 수 있으므로, 같은 image_id로 병합 필요)
                    if not ann_df.empty:
                        merged_df = pd.concat([image_df, etc_df, env_df, growth_df, lic_df, cat_df], axis=1)
                        merged_df = pd.concat([merged_df] * len(ann_df), ignore_index=True)
                        merged_df = pd.concat([merged_df, ann_df.reset_index(drop=True)], axis=1)
                    else:
                        merged_df = pd.concat([image_df, etc_df, env_df, growth_df, lic_df, cat_df], axis=1)
                    
                    # 전체 데이터 프레임에 추가
                    combined_df = pd.concat([combined_df, merged_df], ignore_index=True)

    return combined_df
```
